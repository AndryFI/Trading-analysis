## **Task implementation**

### **Collecting INDEXES Data**

```python
#df1 = pd.read_csv(DATA_PATH + 'panel_data_close.csv', parse_dates=[0], date_parser=parser, decimal=',')
df1 = pd.read_csv(INDEXES_PATH + 'Dates.csv')
df1['Date'] =  pd.to_datetime(df1['Date'], format='%d.%m.%Y')
df1 = df1[df1['Date'] > '2011-04-03']
df1.info()
```

```python
# opening and merging files
filepaths_xlsx_many_cols = []
for filepath in filepaths_xlsx:
    df_x = pd.read_excel(filepath)
    name = filepath.split('/')[-1].split('.')[0]
    if df_x.shape[1] > 2:
        if name.startswith('LIBOR'):
            df_x.drop(['Week day', '1W', '1M', '2M', '3M', '6M', '12M'], axis=1, inplace=True)
            df_x.set_axis(['Date', name], axis = 'columns', inplace = True)
            df_x['Date'] =  pd.to_datetime(df_x['Date'], format='%d.%m.%Y')
            df1 = df1.merge(df_x, on='Date', how='left')
        else:
            filepaths_xlsx_many_cols.append(filepath)
    else:
        df_x.set_axis(['Date', name], axis = 'columns', inplace = True)
        df1 = df1.merge(df_x, on='Date', how='left')
df1.head()
```

```python
df1.info()
```

```python
for filepath in filepaths_csv:
    df_x = pd.read_csv(filepath, sep=',', encoding='latin-1', decimal=',')
    name = filepath.split('/')[-1].split('.')[0]
    if name.startswith('Dates'):
        continue
    else:
        df_x.set_axis(['Date', name], axis = 'columns', inplace = True)
        df_x['Date'] =  pd.to_datetime(df_x['Date'], format='%d.%m.%Y')
        min_date = df_x['Date'].min()
        df1 = df1.merge(df_x, on='Date', how='left')
        df1.loc[df1['Date'] < min_date, name] = 0
df1.head(2)
```

```python
df1.to_csv(DATA_PATH+'Raw_Data_Processing_(INXs)' + REL +'.csv', index=None)
```

### **Collecting BRANCHES Data**

```python
df2 = pd.read_csv(INDEXES_PATH + 'Dates.csv')
df2['Date'] =  pd.to_datetime(df2['Date'], format='%d.%m.%Y')
df2 = df2[df2['Date'] > '2011-04-03']
df2.info()
```

Paths to the branches files

```python
filepaths_branches = [(BRANCHES_PATH + f) for f in os.listdir(BRANCHES_PATH)]
print(len(filepaths_branches))
```

Loading branches data

```python
for filepath in filepaths_branches:
    df_x = pd.read_csv(filepath, parse_dates=[0], date_parser=mgf.parser, decimal=',', thousands='.')
    name = filepath.split('/')[-1].split('.')[0]
    df_x = df_x[['Дата', 'Цена']]
    df_x.set_axis(['Date', name], axis = 'columns', inplace = True)
    df2 = df2.merge(df_x, on='Date', how='left')
df2.head(15)
```

```python
df2.info()
```

```python
df2.to_csv(DATA_PATH+'Raw_Data_Processing_(BRHs)' + REL +'.csv', index=None)
```

### **Saving Data INDEXES + BRANCHES**
